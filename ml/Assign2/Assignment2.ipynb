```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("/content/add.xlsx")
data.head()
```

display summary statistics

```python
data.describe()
```

```python
data.var()
```

histograms of a each feature

```python
chance_of_admit = data['Chance of Admit']
plt.hist(chance_of_admit, edgecolor='black')
plt.title('Chance of Admit Histogram')
plt.xlabel('Chance of Admit')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
gre_scores = data['GRE Score']
plt.hist(gre_scores,edgecolor="black",color="red",alpha=0.7)
plt.title("GRE score")
plt.xlabel("GRE score")
plt.show()
```

```python
ts = data['TOEFL Score']
plt.hist(ts,edgecolor="black",color="m",alpha=0.5)
plt.show()
```

```python
ur = data['University Rating']
plt.hist(ur,edgecolor="black",color="y",alpha=0.5)
plt.title("university rating")
plt.show()
```

```python
s = data['SOP']
plt.hist(s,edgecolor="black",color="m",alpha=0.5)
plt.title("SOP")
plt.show()
```

```python
c=data["CGPA"]
plt.hist(c,edgecolor="r",color='b',linewidth=0.5)
plt.title("cgpa")
plt.show()
```
